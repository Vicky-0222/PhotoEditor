```python
from tkinter import *
from tkinter import ttk
from PIL import ImageTk, Image, ImageFilter, ImageFont, ImageDraw
import random

root = Tk()

# Создание окна
root.title("Фоторедактор")

# Установка размера
root.geometry("700x700")

#Приветствие
welcome = Label(text="Добро пожаловать в фоторедактор")
welcome.pack()

# Загрузка изображения
open_image = Image.open("house.jpg")
my_img = ImageTk.PhotoImage(open_image)
my_label = Label(image=my_img)
my_label.pack()

# Функция для обработки изображения
def apply_filter(filter_name):
    # загрузка изображения в качестве объекта Pillow
    img = open_image
    
    # применение фильтра
    
    #обрезка
    if filter_name == "crop":
        img = img.crop((10,10,100,100))
        filtered_image = img.save("crop.jpg")
    
    elif  filter_name == "rotate_45":
        img = img.rotate(45)
        filtered_image = img.save("rotate_45.jpg")
        
    elif  filter_name == "rotate_90":
        img = img.rotate(90)
        filtered_image = img.save("rotate_90.jpg")
    
    elif  filter_name == "rotate_180":
        img = img.rotate(100)
        filtered_image = img.save("rotate_180.jpg")
        
    elif  filter_name == "rotate_270":
        img = img.rotate(270)
        filtered_image = img.save("rotate_270.jpg")
    
    
    #отразить зеркально по горизонтали(слева направо) 
    elif filter_name == "transpose_left_right":
        img  = img.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
        filtered_image = img.save("trans_l_r.jpg")
     
    #отразить зеркально по вертикали(сверху вниз) 
    elif filter_name == "transpose_top_bottom":
        img = img.transpose(Image.Transpose.FLIP_TOP_BOTTOM)
        filtered_image = img.save("trans_l_r.jpg")
     
    #резкость
    elif filter_name == "detail":
        img = img.filter(ImageFilter.DETAIL)
        filtered_image = img.save("detail.jpg")
     
    #размытие
    elif filter_name == "blur":
        img = img.filter(ImageFilter.BLUR)
        filtered_image = img.save("blur.jpg")
     
    #цветовой шум
    elif filter_name == "color_noise":
        for i in range(img.width):
            for j in range(img.height):
                r, g, b = img.getpixel((i, j))
                r += random.randint(-150, 150)
                g += random.randint(-150, 150)
                b += random.randint(-150, 150)
                img.putpixel((i, j), (r, g, b))
        filtered_image = img.save("color.jpg")
    
    #оттенки серого
    elif filter_name == "gray":
        img = img.convert('L')
        filtered_image = img.save("gray.jpg")
        
    #сепия
    elif filter_name == "sepia":
        sepia = lambda r,g,b: (int(r * 0.393 + g * 0.769 + b * 0.189),
                              int(r * 0.349 + g * 0.686 + b * 0.168),
                              int(r * 0.272 + g * 0.534 + b * 0.131))
        img = img.convert('RGB')        
        img = img.point(sepia)
        filtered_image = img.save("sepia.jpg")
     
    #наложение текста
    elif filter_name == "text":
        font = ImageFont.load_default()
        pencil = ImageDraw.Draw(img)
        pencil.text((50,50), "Hello world", font = font, fill = "red", size=100)
        filtered_image = img.save("text.jpg")
        

    
    
# Создание кнопок
crop_btn = Button(root, text="Обрезка", command=lambda: apply_filter("crop"))
crop_btn.pack()

rotate_btn = Button(root, text="Поворот_45", command=lambda: apply_filter("rotate_45"))
rotate_btn.pack()

rotate_btn = Button(root, text="Поворот_90", command=lambda: apply_filter("rotate_90"))
rotate_btn.pack()

rotate_btn = Button(root, text="Поворот_180", command=lambda: apply_filter("rotate_180"))
rotate_btn.pack()

rotate_btn = Button(root, text="Поворот_270", command=lambda: apply_filter("rotate_270"))
rotate_btn.pack()

trans_l_r_btn = Button(root, text="Отразить зеркально по горизонтали(слева направо)", command=lambda: apply_filter("transpose_left_right"))
trans_l_r_btn.pack()

trans_s_b_btn = Button(root, text="Отразить зеркально по вертикали(сверху вниз)", command=lambda: apply_filter("transpose_top_bottom"))
trans_s_b_btn.pack()

detail_btn = Button(root, text="Резкость", command=lambda: apply_filter("detail"))
detail_btn.pack()

blur_btn = Button(root, text="Размытие", command=lambda: apply_filter("blur"))
blur_btn.pack()

color_noise_btn = Button(root, text="Цветовой шум", command=lambda: apply_filter("color_noise"))
color_noise_btn.pack()

gray_btn = Button(root, text="Оттенки серого", command=lambda: apply_filter("gray"))
gray_btn.pack()

sepia_btn = Button(root, text="Сепия", command=lambda: apply_filter("sepia"))
sepia_btn.pack()

text_btn = Button(root, text="Наложение текста", command=lambda: apply_filter("text"))
text_btn.pack()

# Запуск приложения
root.mainloop()
```
